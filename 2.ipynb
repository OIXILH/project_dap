```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
occupation= pd.read_csv("https://raw.githubusercontent.com/OIXILH/project_dap/main/u.occupation")
user= pd.read_csv("https://raw.githubusercontent.com/OIXILH/project_dap/main/u.user",delimiter='|')
data= pd.read_csv("https://raw.githubusercontent.com/OIXILH/project_dap/main/u.data",delimiter='\t')
genre= pd.read_csv("https://raw.githubusercontent.com/OIXILH/project_dap/main/u.genre",delimiter='|')
item= pd.read_csv("https://raw.githubusercontent.com/OIXILH/project_dap/main/u.item",delimiter='|')
```

```python
result_totaluser=pd.merge(data,user,how="inner")
```

```python
result_gender=pd.merge(data,user,how="inner")
result_gender=result_gender[result_gender['gender']=='M']
total_result_M=pd.merge(result_gender,item,how="inner")
total_result_M=total_result_M.drop(['user_id','rating','item_id','movie title','age','gender','occupation','timestamp','zipcode','release date','video release date','IMDb URL'], axis=1)
sums_M=total_result_M.sum()
sums_M=sums_M.sort_values(ascending = False)
result_totalM=pd.merge(data,user,how="inner")
result_totalM=result_totalM[result_totalM['gender']=='M']
agv_result_totalM=pd.merge(result_totalM,item,how="inner")
total_result_drop_M=agv_result_totalM.drop(['rating','user_id','item_id','movie title','age','gender','occupation','timestamp','zipcode','release date','video release date','IMDb URL'], axis=1)
total_result_drop_M['unknown']=total_result_drop_M['unknown']*agv_result_totalM['rating']
total_result_drop_M['Action']=total_result_drop_M['Action']*agv_result_totalM['rating']
total_result_drop_M['Adventure']=total_result_drop_M['Adventure']*agv_result_totalM['rating']
total_result_drop_M['Animation']=total_result_drop_M['Animation']*agv_result_totalM['rating']
total_result_drop_M["Children's"]=total_result_drop_M["Children's"]*agv_result_totalM['rating']
total_result_drop_M['Comedy']=total_result_drop_M['Comedy']*agv_result_totalM['rating']
total_result_drop_M['Crime']=total_result_drop_M['Crime']*agv_result_totalM['rating']
total_result_drop_M['Documentary']=total_result_drop_M['Documentary']*agv_result_totalM['rating']
total_result_drop_M['Drama']=total_result_drop_M['Drama']*agv_result_totalM['rating']
total_result_drop_M['Fantasy']=total_result_drop_M['Fantasy']*agv_result_totalM['rating']
total_result_drop_M['Film-Noir']=total_result_drop_M['Film-Noir']*agv_result_totalM['rating']
total_result_drop_M['Horror']=total_result_drop_M['Horror']*agv_result_totalM['rating']
total_result_drop_M['Musical']=total_result_drop_M['Musical']*agv_result_totalM['rating']
total_result_drop_M['Mystery']=total_result_drop_M['Mystery']*agv_result_totalM['rating']
total_result_drop_M['Romance']=total_result_drop_M['Romance']*agv_result_totalM['rating']
total_result_drop_M['Sci-Fi']=total_result_drop_M['Sci-Fi']*agv_result_totalM['rating']
total_result_drop_M['Thriller']=total_result_drop_M['Thriller']*agv_result_totalM['rating']
total_result_drop_M['War']=total_result_drop_M['War']*agv_result_totalM['rating']
total_result_drop_M['Western']=total_result_drop_M['Western']*agv_result_totalM['rating']
agv_sums_M=total_result_drop_M.sum()
agv_sums_M=agv_sums_M/sums_M
agv_sums_M=agv_sums_M.sort_values(ascending = False)
agv_sums_M=agv_sums_M.to_frame(name='avg')
agv_sums_M#남자사용자 장르 평점의 평균나열
```
